```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
df = pd.read_csv('./diabetes.csv')
```

```python
df.head()

type(df)
```

```python
x = df
y = df.pop('Outcome')

y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
y_train
```

```python
model = DecisionTreeClassifier(max_depth = 2)
model = model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
plot_tree(model)
```

```python
diff = y_predict - np.array(y_test)
print('number of incorrect classifications are ', len(diff) - list(diff).count(0), 'out of ', len(diff))
print('accuracy = ', list(diff).count(0)/len(diff)*100, '%')
```
