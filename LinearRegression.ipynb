```python
import numpy as np
import pandas as pd


df = pd.read_csv(
    'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')

df.head()
```

```python
df = df[['petal.length','petal.width']]

df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df.pop('petal.width')

```

```python
df.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(df,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)

```

```python
y_predict = lr.predict(x_test)
y_real = np.array(y_test)
```

```python
from sklearn.metrics import mean_squared_error

print('Root Mean Square = ',mean_squared_error(y_predict,y_real)**0.5)
```

```python
import matplotlib.pyplot as pt

pt.scatter(x_test,y_test)
pt.scatter(x_test, y_predict)
pt.xlabel('petal Length')
pt.ylabel('petal width')
```
